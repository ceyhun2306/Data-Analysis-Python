#Gender Distribution and Age Buckets

```python
bins = pd.cut(df['Age'], bins=[0,18,20,22,25,100], labels=['<=18','19-20','21-22','23-25','26+'])

fig, axes = plt.subplots(1, 2, figsize=(12,5))

df['Gender'].value_counts().plot(kind='bar', ax=axes[0], color=['skyblue', 'lightgreen'])
axes[0].set_title('Gender Distribution')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Count')

bins.value_counts().sort_index().plot(kind='barh', ax=axes[1], color='salmon')
axes[1].set_title('Age Buckets')
axes[1].set_xlabel('Count')
axes[1].set_ylabel('Age Bucket')

plt.show()
```

In the first visual, we observe the difference between genders. Although the gap isn’t very large, the number of males slightly exceeds that of females. There are over 100 male students, while females are close to 80, meaning males make up roughly 25% more of the group.
In the second visual, we see that most students are between the ages of 21 and 22. This suggests that students in this age range are more active and eager to work on new projects. The 19–20 and 23–25 age groups have nearly equal representation, while students aged 26 and above are fewer than 20, and there are almost no students under 18. Overall, this indicates that students aged 21–22 are the most engaged group in such projects.


#SQL, Python, And Java Levels Of Students

```python
top_maj = df['Major'].value_counts().head(8).index

fig, axes = plt.subplots(1, 3, figsize=(15,5))

ct_sql = pd.crosstab(df.loc[df['Major'].isin(top_maj), 'Major'], df['SQL']).reindex(columns=['Strong','Average','Weak'])
ct_sql.plot(kind='bar', stacked=True, ax=axes[0], color=['skyblue','orange','lightgreen'])
axes[0].set_title('SQL Levels')
axes[0].set_xlabel('Major')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=360)

ct_py = pd.crosstab(df.loc[df['Major'].isin(top_maj), 'Major'], df['Python']).reindex(columns=['Strong','Average','Weak'])
ct_py.plot(kind='bar', stacked=True, ax=axes[1], color=['skyblue','orange','lightgreen'])
axes[1].set_title('Python Levels')
axes[1].set_xlabel('Major')
axes[1].set_ylabel('')
axes[1].tick_params(axis='x', rotation=360)

ct_java = pd.crosstab(df.loc[df['Major'].isin(top_maj), 'Major'], df['Java']).reindex(columns=['Strong','Average','Weak'])
ct_java.plot(kind='bar', stacked=True, ax=axes[2], color=['skyblue','orange','lightgreen'])
axes[2].set_title('Java Levels')
axes[2].set_xlabel('Major')
axes[2].set_ylabel('')
axes[2].tick_params(axis='x', rotation=360)

plt.tight_layout()
plt.show()
```

When we look at these visuals, we can see how students differ from one another in terms of their SQL, Python, and Java skills.

Starting with **SQL**, the number of *Strong* students is noticeably higher compared to the others. Similarly, the number of *Average* students is also greater than that of *Weak* students. This indicates that most students are quite skilled and successful in SQL.

Moving on to **Python**, the results show that students tend to perform slightly weaker compared to SQL. Both the *Average* and *Strong* levels are lower than in SQL, suggesting that Python is a bit more challenging for them.

The same applies to **Java** — students’ performance is similar to Python, with a higher number of *Weak* learners compared to SQL. The counts of *Average* and *Strong* students in Java are nearly the same as those in Python.

Overall, this shows that students are generally more proficient in **SQL**, while their performance in **Python** and **Java** is relatively close but slightly weaker.


#Top Future Career Choices (Top 10)

```python
fc = df['Future Career'].value_counts().head(10).sort_values()
bars = plt.barh(fc.index, fc.values)
plt.title('Top Future Career Choices (Top 10)')
plt.xlabel('Count'); plt.ylabel('Career')
plt.bar_label(bars)
plt.show()
```

This visual represents the **Top 10 Future Career Choices** among students.
As we can see, **Web Developer** is the most preferred career path, followed closely by **Mobile App Developer** and **Information Security Analyst**. These three fields seem to attract the largest number of students, reflecting the growing demand for web, mobile, and cybersecurity specialists in the tech industry.

In contrast, careers such as **Graphics Programmer** and **Data Scientist** are less common choices among students, though they remain highly valuable and specialized fields.

Overall, the chart suggests that students are primarily interested in **development-focused** and **security-related** roles, showing a strong inclination toward practical, applied areas of computer science.

#Average GPA by Future Career (Top 10)

```python
gpa_by_career = (df.groupby('Future Career', as_index=False)['GPA'].mean().sort_values(by='GPA', ascending=False))
plt.figure(figsize=(9,5))
bars = plt.barh(gpa_by_career['Future Career'].head(10), gpa_by_career['GPA'].head(10), color='skyblue')
plt.title('Average GPA by Future Career (Top 10)')
plt.xlabel('Average GPA')
plt.ylabel('Future Career')
plt.bar_label(bars)
plt.gca().invert_yaxis()
plt.show()
```

This visual shows the relationship between **GPA** and **Future Career** choices.
It reveals that students aiming to become **Distributed Systems Engineers**, **NLP Engineers**, and **Quantum Computing Researchers** have the highest average GPAs.
Overall, students with stronger academic performance tend to prefer **advanced and research-oriented careers** such as AI, data science, and engineering fields.


#Relationship Between Age and GPA

```python
gpa_by_age = df.groupby('Age', as_index=False)['GPA'].mean()

plt.figure(figsize=(8,5))
plt.plot(gpa_by_age['Age'], gpa_by_age['GPA'], marker='o', color='orange')
plt.title('Relationship Between Age and GPA')
plt.xlabel('Age')
plt.ylabel('Average GPA')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

This visual illustrates the relationship between Age and GPA.
It shows that students aged 20 to 23 generally maintain a consistent GPA around 3.6, while older students, especially those aged 34–35, achieve slightly higher GPAs, reaching nearly 3.9.
Overall, GPA remains stable across most ages, but mature students tend to perform a bit better academically.
